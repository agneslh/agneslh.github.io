```python
wbd_dir = et.data.get_data(
    url='https://prd-tnm.s3.amazonaws.com/' 
    'StagedProducts/Hydrography/WBD/National/GDB/WBD_National_GDB.zip')
wbd_dir
```

```python
wbd_gdb = os.path.join(wbd_dir, 'WBD_National_GDB.gdb')

wbdhu4_gdf = gpd.read_file(wbd_gdb,driver='OpenFileGBD',layer='WBDHU4',from_disk=True)
wbdhu4_gdf
```

```python
usfs_dir = et.data.get_data(
    url = 'https://www.fs.usda.gov/rds/archive/products/RDS-2013-0009.6'
    '/RDS-2013-0009.6_Data_Format4_SQLITE.zip'
)
usfs_dir
```

```python
if not 'fire_occurrence_df' in globals():
    fire_db_path =  os.path.join(usfs_dir, "Data", "FPA_FOD_20221014.sqlite")
    fire_occurrence_df = pd.read_sql_query(
        "SELECT * FROM Fires",
       "sqlite:///" + fire_db_path)

fire_occurrence_df.head()
```

```python
cleaned_df = fire_occurrence_df[['FOD_ID', 'LATITUDE', 'LONGITUDE', 'DISCOVERY_DATE', 
    'NWCG_GENERAL_CAUSE', 'FIRE_SIZE']]

cleaned_df.set_index('FOD_ID', inplace= True)

cleaned_df['DISCOVERY_DATE'] = pd.to_datetime(cleaned_df['DISCOVERY_DATE'])

cleaned_gdf = gpd.GeoDataFrame(
    cleaned_df, 
    geometry=gpd.points_from_xy(cleaned_df.LONGITUDE, cleaned_df.LATITUDE), 
    crs="EPSG:4326"
)
wbdhu4_gdf.crs
cleaned_gdf = cleaned_gdf.to_crs({'init': 'epsg:5498'})

cleaned_gdf
```

```python
fire_region_gdf = (
    wbdhu4_gdf
    [['name', 'geometry']]
    .sjoin(cleaned_gdf, how='inner', predicate='intersects')
)

fire_region_gdf = (
    fire_region_gdf.groupby
    ([ fire_region_gdf.DISCOVERY_DATE.dt.year, 'name'])
    .agg(mean_fire_size=('FIRE_SIZE', 'mean'), num_fires=('index_right', 'count'))
    .reset_index()
)

#try switching the 'name' groupby and including geometry instead of index_right
fire_region_gdf
```

```python
# Identify the unique years
unique_years = fire_region_gdf['DISCOVERY_DATE'].unique()
unique_years.sort()

# Get the first and last five years
first_five_years = unique_years[:5]
last_five_years = unique_years[-5:]

# # Calculate mean fire size for the first five years
first_five_mean_size = fire_region_gdf[fire_region_gdf['DISCOVERY_DATE'].isin(first_five_years)]['mean_fire_size'].mean()

# # Calculate mean fire size for the last five years
last_five_mean_size = fire_region_gdf[fire_region_gdf['DISCOVERY_DATE'].isin(last_five_years)]['mean_fire_size'].mean()

# # Add these to fire_region_gdf as new columns
fire_region_gdf['avg_fire_size_first_5_years'] = first_five_mean_size
fire_region_gdf['avg_fire_size_last_5_years'] = last_five_mean_size

avg_fire_size_first_five_years_gdf = (fire_region_gdf[
    fire_region_gdf['DISCOVERY_DATE'].isin(first_five_years)]
    .groupby(['name'])
    .agg(avg_first_five_years_fire_size=('mean_fire_size', 'mean'))
    .reset_index())

avg_fire_size_last_five_years_gdf = (fire_region_gdf[
    fire_region_gdf['DISCOVERY_DATE'].isin(last_five_years)]
    .groupby(['name'])
    .agg(avg_last_five_years_fire_size=('mean_fire_size', 'mean'))
    .reset_index())
# fire_region_gdf[(fire_region_gdf['DISCOVERY_DATE'].isin(first_five_years)) & (fire_region_gdf['name'] == 'Alabama')]
# fire_region_gdf[fire_region_gdf['DISCOVERY_DATE'].isin(first_five_years)]
avg_fire_size_df = avg_fire_size_first_five_years_gdf.merge(avg_fire_size_last_five_years_gdf, on="name")
avg_fire_size_df['fire_size_difference'] = avg_fire_size_df['avg_last_five_years_fire_size'] - avg_fire_size_df['avg_first_five_years_fire_size']
avg_fire_size_df
```

```python
# Get the unique years
unique_years = fire_region_gdf['DISCOVERY_DATE'].unique()
unique_years.sort()

# Get the first and last five years labels
first_five_years_label = f"{unique_years[0]}-{unique_years[4]}"
last_five_years_label = f"{unique_years[-5]}-{unique_years[-1]}"

# Data for plotting
labels = [first_five_years_label, last_five_years_label, 'Delta']
values_first = fire_region_gdf['avg_fire_size_first_5_years'].iloc[0]
values_last = fire_region_gdf['avg_fire_size_last_5_years'].iloc[0]
delta = abs(values_last - values_first)
values = [values_first, values_last, delta]

# Create the bar plot
colors = ['blue', 'red', 'green']
bars = plt.bar(labels, values, color=colors)

# Add legend
legend_labels = ['First 5 Years of Data', 'Last 5 Years of Data', 'Delta']
plt.legend(bars, legend_labels)

plt.xlabel('Year Range')
plt.ylabel('Average Fire Size')
plt.title('Change in Average Fire Size Over Time')
plt.show()
```

```python
# Calculate the difference between 'avg_fire_size_last_5_years' and 'avg_fire_size_first_5_years' for each 'name'
fire_region_gdf['fire_size_difference'] = fire_region_gdf['avg_fire_size_last_5_years'] - fire_region_gdf['avg_fire_size_first_5_years']

# Group by 'name' and sum the differences
grouped_difference = avg_fire_size_df.groupby('name')['fire_size_difference'].sum()

# SelectMultiple widget
multi_select = widgets.SelectMultiple(
    options=grouped_difference.index.tolist(),
    value=[grouped_difference.index[0]],
    description='HUC4 Areas:',
    disabled=False
)

# Function to plot when regions are selected
def plot_regions(change):
    clear_output(wait=True)
    display(multi_select)
    
    selected_regions = list(change['new'])
    
    # Filter data to include only selected regions
    filtered_data = grouped_difference[grouped_difference.index.isin(selected_regions)]
    
    filtered_data.sort_values().plot(kind='bar', figsize=(12, 6), color='skyblue')
    plt.ylabel('Difference in Average Fire Size')
    plt.xlabel('HUC4 Region')
    plt.title('Difference in Average Fire Size from First 5 Years to Last 5 Years')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

multi_select.observe(plot_regions, names='value')
display(multi_select)
```

```python
fire_region_gdf['fire_size_difference'] = fire_region_gdf['avg_fire_size_last_5_years'] - fire_region_gdf['avg_fire_size_first_5_years']

# Merge the 'fire_size_difference' back to your original GeoDataFrame (wbdhu4_gdf) to get the geometry for each region
merged_gdf = wbdhu4_gdf.merge(avg_fire_size_df, on='name', how='left')

merged_gdf = merged_gdf[['name', 'fire_size_difference', 'geometry']].drop_duplicates()
merged_gdf.geometry = merged_gdf.geometry.simplify(tolerance=.1)
merged_gdf
```
